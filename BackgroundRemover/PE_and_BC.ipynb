```python
import cv2
import cvzone
from cvzone.SelfiSegmentationModule import SelfiSegmentation
import os
from cvzone.PoseModule import PoseDetector
```

```python
cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)
# cap.set(cv2.CAP_PROP_FPS, 60)
```

```python
segmentor = SelfiSegmentation()
fpsReader = cvzone.FPS()

# imgBG = cv2.imread("BackgroundImages/3.jpg")

listImg = os.listdir("bgimages")
imgList = []
for imgPath in listImg:
    img = cv2.imread(f'bgimages/{imgPath}')
    imgList.append(img)

indexImg = 0
```

```python
detector = PoseDetector()
```

```python
while True:
    success, img = cap.read()
    img = detector.findPose(img)
    lmList, bboxInfo = detector.findPosition(img, bboxWithHands=False)
    # imgOut = segmentor.removeBG(img, (255,0,255), threshold=0.83)
    imgOut = segmentor.removeBG(img, imgList[indexImg], threshold=0.8)

    imgStack = cvzone.stackImages([img, imgOut], 2,1)
    _, imgStack = fpsReader.update(imgStack)
    print(indexImg)
    
    if bboxInfo:
        center = bboxInfo["center"]
        cv2.circle(img, center, 5, (255, 0, 255), cv2.FILLED)
        
    cv2.imshow("image", imgStack)
    key = cv2.waitKey(1)
    if key == ord('a'):
        indexImg +=1
        if indexImg>=len(imgList):
            indexImg = 0
    elif key == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
